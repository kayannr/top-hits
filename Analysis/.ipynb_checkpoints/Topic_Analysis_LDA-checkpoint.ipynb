# Topic Analysis

## I. Latent Dirichlet Allocation (LDA) <a class="anchor" id="lda"></a>

```python
# IMPORT DEPENDENCIES 
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import multiprocessing
import pyLDAvis.gensim_models 
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
import gensim.corpora as corpora
from pprint import pprint
import pickle 
import pyLDAvis
import time
import os
from gensim.models import LdaMulticore
import matplotlib.pyplot as plt
import seaborn as sns
from sentence_transformers import SentenceTransformer
from umap import UMAP
from sklearn.feature_extraction.text import CountVectorizer
from gensim.models import CoherenceModel
sns.set()
```

```python
# READ DATA 
song_data = pd.read_csv('../Data/merged_finaltop100_revised.csv') 
# EXCLUDE ROWS WITH NULL VALUES 
song_data = song_data.dropna()
song_data
```

```python
# GET NUMBER OF UNIQUE TOKENS 
text = ' '.join(song_data['lyrics_clean']).split() 
new_set = set(text)
print("# of unique tokens:", len(new_set))
```

```python
# TOKENIZE TEXT 
tokenized_lyrics = [nltk.word_tokenize(text) for text in song_data['lyrics_clean']]

# CONVERT TOKENIZED OBJECT INTO CORPUS AND DICTIONARY 
id2word  = corpora.Dictionary(tokenized_lyrics) #create dict 
print(id2word.token2id) 
```

```python
# FILTER OUT VERY LOW-FREQUENCY AND HIHG-FREQUENCY TOKENS 
id2word.filter_extremes(no_below=5, no_above=0.5, keep_n=1000)
print(len(id2word.token2id))
```

```python
# TERM DOCUMENT FREQUENCY 
corpus = [id2word.doc2bow(text) for text in tokenized_lyrics]
print(corpus[:1][0][:20]) #view 
```

**Choose optimal topic size**

```python
%%time 
# CHOOSE OPTIMAL TOPIC SIZE USINF UMASS 
topics = []
score = []
for i in range(1,20,1):
    lda_model = LdaMulticore(corpus=corpus, id2word=id2word, iterations=10, num_topics=i, workers = 4, passes=10, random_state=100)
    cm = CoherenceModel(model=lda_model, corpus=corpus, dictionary=id2word, coherence='u_mass')
    topics.append(i)
    score.append(cm.get_coherence())
plt.plot(topics, score)
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Score (u_mass)')
plt.title('Figure 5.1: Number of topics vs. Coherence score')
plt.show()
```

It appears that the coherence score generally decreases as the number of topics increases. U_Mass calculates how often two words appear together, and averages the pairwise coherence scores over the top N words for each topic. For UMass, the higher the score, the better the performance. 

```python
%%time 
# CHOOSE OPTIMAL TOPIC SIZE USINF CV 
topics = []
score = []
for i in range(1,20,1):
    lda_model = LdaMulticore(corpus=corpus, id2word=id2word, iterations=10, num_topics=i, workers = 4, passes=10, random_state=100)
    cm = CoherenceModel(model=lda_model, texts = tokenized_lyrics, corpus=corpus, dictionary=id2word, coherence='c_v')
    topics.append(i)
    score.append(cm.get_coherence())
plt.plot(topics, score)
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Score (u_mass)')
plt.title('Figure 5.2: Number of topics vs. Coherence score')
plt.show()
```

Unlike U_mass, CV scores also generally fluctuates as topic size changes. CV measures how often topic words appear together in the corpus. It creates content vectors of words using their co-occurrences and then computes the score using NPMI and the cosine similarity.

**Fit LDA with optimal topic size**

For this model, I choose the optimal topic size as 6 with moderate u_mass and cv scores.  

```python
%%time
# FIT LDA WITH OPTIMAL TOPIC SIZE 
lda_model = gensim.models.LdaMulticore(corpus=corpus,id2word=id2word,num_topics=6)
```

```python
# PRINT RESULTS 
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
# PRINT FIRST 4 TOPICS AND DOMINANT WORDS 
for topic_id, topic in lda_model.print_topics(num_topics=6):
    print(f"Topic {topic_id}: {topic}")
```

```python
# VISUALIZE RESULTS 
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary=lda_model.id2word)
vis
```

*Figure 5.3: LDA Plots*

It seems that Topic 6 represents a topic related to night life since it includes words such as part, dance, and night. 

```python
# COMPUTE COHERENCE SCORE 
from gensim.models.coherencemodel import CoherenceModel
coherence_model_lda = CoherenceModel(model=lda_model, texts=tokenized_lyrics, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('nCoherence Score: ', coherence_lda)
```

It appears that unsupervised topic modelling does not seem to work very well with song lyrics data. One possible reason could be the size of the corpus.  A small document corpus may not contain enough information to create distinct topics that are coherent. Increasing the size of the corpus may help improve the coherence score.
